# 601139 基本面分析

```python
#导入包
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.sans-serif'] = ['arial unicode ms']
mpl.rcParams['axes.unicode_minus'] = False

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
#获取数据
import tushare as ts
pro = ts.pro_api('0f56d40cab4fcd3fb4651d1a9538c42007971f09ceb90cbaaa9e2969')
df = pro.balancesheet(ts_code='601139.SH')

# 删除同样的日期
a = df.end_date[:-1].copy()
a.index = np.arange(1,len(df.index))
a = a.reindex(np.arange(len(df.index)))

df['bool'] = a == df.end_date
c = df[df['bool'] == False]
c.index = np.arange(len(c))

c['quar'] = 'a'
for i in np.arange(len(c)):
    c['quar'][i] = str(c.end_date[i])[5]

c.to_csv('601139.SH.csv')   
```

## 资产负债表

```python
#导入数据
df = pd.read_csv('601139.SH.csv')
```

## 总资产

```python
#数据分类
first_q = df[df.quar == 3]
second_q = df[df.quar == 6]
third_q = df[df.quar == 9]
fourth_q = df[df.quar == 2]

q_1 = first_q.total_assets.values
q_2 = second_q.total_assets.values
q_3 = third_q.total_assets.values
q_4 = fourth_q.total_assets.values

year_1 = first_q.end_date.values//10000
year_2 = second_q.end_date.values//10000
year_3 = third_q.end_date.values//10000
year_4 = fourth_q.end_date.values//10000

# 总资产图表

width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(year_1-0.4, q_1,width,align='edge', label='一季度')
rects2 = ax.bar(year_2 - 0.2, q_2, width, align='edge',label='二季度')
rects3 = ax.bar(year_3 , q_3, width, align='edge',label='三季度')
rects4 = ax.bar(year_4 + 0.2, q_4, width, align='edge',label='四季度')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('总资产')
ax.set_title('总资产时间图')
ax.set_xticks(year_4)

ax.legend()


fig.tight_layout()

plt.show()
```

```python
# 总资产增长率
dq_1 = (q_1[:-1] - q_1[1:])/q_1[1:]
dq_2 = (q_2[:-1] - q_2[1:])/q_2[1:]
dq_3 = (q_3[:-1] - q_3[1:])/q_3[1:]
dq_4 = (q_4[:-1] - q_4[1:])/q_4[1:]
#ddq_1 = (dq_1[:-1] - dq_1[1:])/dq_1[1:]
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(year_1[:-1]-0.4, dq_1,width,align='edge', label='一季度')
rects2 = ax.bar(year_2[:-1] - 0.2, dq_2, width, align='edge',label='二季度')
rects3 = ax.bar(year_3[:-1] , dq_3, width, align='edge',label='三季度')
rects4 = ax.bar(year_4[:-1] + 0.2, dq_4, width, align='edge',label='四季度')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('总资产增长率')
ax.set_title('总资产增长率时间图')
ax.set_xticks(year_4)

ax.legend()


fig.tight_layout()

plt.show()


```
